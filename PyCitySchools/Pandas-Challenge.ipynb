```python
import pandas as pd
```

```python
schoolfile = "../Resources/schools_complete.csv"
studentsfile = "../Resources/students_complete.csv"
```

```python
schools_df = pd.read_csv(schoolfile)
students_df = pd.read_csv(studentsfile)
```

```python
# merge to one data set
Full_Data_df = pd.merge(schools_df,students_df,how = "left", on = "school_name")
Full_Data_df.head()
```

```python
#District Summary
#Total Schools
Schools = Full_Data_df["school_name"].unique()
```

```python
Total_Schools = len(Schools)
Total_Schools
```

```python
#Total Students
Students = Full_Data_df["student_name"].unique()
Total_Students = len(Students)
Total_Students
```

```python
#Total Budget
school_budgets = schools_df["budget"].sum()
school_budgets
```

```python
#Average Maths Score
Avg_math = Full_Data_df["math_score"].mean()
Avg_math
```

```python
#Average Reading Score
Avg_read = Full_Data_df["reading_score"].mean()
Avg_read
```

```python
#% passing math (the percentage of students who passed math)
def math_result(row):
    if row["math_score"] < 70:
        return "Failed"
    else:
        return "Passed"
```

```python
#Add Pass/Fail Column
Full_Data_df["math_result"] = Full_Data_df.apply(math_result, axis = 1)
Full_Data_df.head()
```

```python
math_p_count = Full_Data_df["math_result"].value_counts()["Passed"]
math_p_rate = (math_p_count / Full_Data_df["math_score"].count()) * 100
math_p_rate
```

```python
#% passing reading (the percentage of students who passed reading)
def read_result(row):
    if row["reading_score"] < 70:
        return "Failed"
    else:
        return "Passed"
```

```python
Full_Data_df["reading_result"] = Full_Data_df.apply(read_result, axis = 1)
Full_Data_df.head()
```

```python
read_p_count = Full_Data_df["reading_result"].value_counts()["Passed"]
read_p_rate = (read_p_count / Full_Data_df["reading_score"].count())*100
read_p_rate
```

```python
#% overall passing (the percentage of students who passed math AND reading)
def Passed_Both (row):
    if row["reading_result"] == "Passed" and row["math_result"] == "Passed":
        return "Yes"
    else:
        return "No"
```

```python
Full_Data_df["Passed_Both"] = Full_Data_df.apply(Passed_Both, axis = 1)
Full_Data_df.head()
```

```python
overall_p_count = Full_Data_df["Passed_Both"].value_counts()["Yes"]
overall_p_rate = (overall_p_count / Full_Data_df["Passed_Both"].count())*100
overall_p_rate
```

```python
#School type
school_type = schools_df.set_index(["school_name"])["type"]
school_type
```

```python
#combuune dataframes
merged_df = pd.merge(schools_df,students_df,on="school_name")
merged_df
```

```python
#number of students
number_students = merged_df["school_name"].value_counts()
number_students
```

```python
#school budget
school_budget = merged_df.groupby(["school_name"]).mean()["budget"]
school_budget
```

```python
#budget per student
bps = school_budget / number_students
bps
```

```python
#average math scores
avg_math = merged_df.groupby(["school_name"]).mean()["math_score"]
avg_math
```

```python
#average reading scores
avg_read = merged_df.groupby(["school_name"]).mean()["reading_score"]
avg_read
```

```python
#Pass Math %
pass_math = merged_df[(merged_df["math_score"]>70)]
pass_math_school = pass_math.groupby(["school_name"]).count()["student_name"]

math_pp = (pass_math_school / number_students) *100

math_pp
```

```python
#Pass Reading %
pass_read = merged_df[(merged_df["reading_score"]>70)]
pass_read_school = pass_read.groupby(["school_name"]).count()["student_name"]

read_pp = (pass_read_school / number_students) *100

read_pp
```

```python
overall_pp = (read_pp + math_pp) / 2 
overall_pp
```

```python
school_summary = pd.DataFrame({"School Type": school_type,
                              "Total Students": number_students,
                              "Total School Budget": school_budget,
                              "Per Student Budget": bps,
                              "Average Math Score": avg_math,
                              "Average Reading Score": avg_read,
                              "% Passing Math": math_pp,
                              "% Passing Reading": read_pp,
                              "Overall Passing Rate": overall_pp})
school_summary
```

```python
school_summary["Total School Budget"] = school_summary["Total School Budget"].map("${:,.2f}".format)
school_summary["Per Student Budget"] = school_summary["Per Student Budget"].map("${:,.2f}".format)
school_summary["Average Math Score"] = school_summary["Average Math Score"].map("{:.2f}".format)
school_summary["Average Reading Score"] = school_summary["Average Reading Score"].map("{:.2f}".format)
school_summary["% Passing Math"] = school_summary["% Passing Math"].map("{:.2f}".format)
school_summary["% Passing Reading"] = school_summary["% Passing Reading"].map("{:.2f}".format)
school_summary["Overall Passing Rate"] = school_summary["Overall Passing Rate"].map("{:.2f}".format)

school_summary.rename(columns={"Overall Passing Rate" : "% Passing Overall"}, inplace = True)

school_summary
```

```python
top_five_sch = school_summary.sort_values("% Passing Overall", ascending = False).head()
top_five_sch
```

```python
bottom_five_sch = school_summary.sort_values("% Passing Overall", ascending = True).head()
bottom_five_sch
```

## Scores by School Spending

```python
#Create a table that breaks down school performance based on average spending ranges (per student). Use your judgment to create 
#four bins with reasonable cutoff values to group school spending. Include the following metrics in the table:

#Average math score
#Average reading score
#% passing math (the percentage of students who passed math)
#% passing reading (the percentage of students who passed reading)
#% overall passing (the percentage of students who passed math AND reading)
```

```python
spend_df = pd.DataFrame({"School Name":Schools,
                         "Budget Per Student": bps,
                        "Average Math Score": avg_math ,
                        "Average Reading Score": avg_read,
                        "Math Pass %": math_pp,
                        "Reading Pass %": read_pp,
                        "Overall Pass %": overall_pp
                        })
spend_df
```

```python
spend_df.reset_index(drop=True, inplace = True)
```

```python
spend_df["Budget Per Student"].max()
```

```python
spend_df["Budget Per Student"].min()
```

```python
bins = [0.00, 599.99,624.99,649.99,675]

group_names = ["Low Spend", "Medium Spend", "High Spend", "Very High Spend"]
```

```python
spend_df["Spending Band"] = pd.cut(bps, bins, labels=group_names, include_lowest=True)
spend_df
```

```python
Spend_Bands = spend_df.groupby("Spending Band")
Spend_Bands.mean()
```

This data shows that lower budget schools in this district are performing better with higher pass rates in both Math and Reading vs schools with higher budgets

```python
size_df = pd.DataFrame({"Number of Students": number_students,
                        "Average Math Score": avg_math ,
                        "Average Reading Score": avg_read,
                        "Math Pass %": math_pp,
                        "Reading Pass %": read_pp,
                        "Overall Pass %": overall_pp
                        })
size_df
```

Data appears to show that smaller schools have a better average score on both maths and reading. They also have a better pass rate that large schools in both subjects.
