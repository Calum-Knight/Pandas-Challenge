```python
# merge to one data set
Full_Data_df = pd.merge(schools_df,students_df,how = "left", on = "school_name")
Full_Data_df.head()
```

```python
#Average Maths Score
Avg_math = Full_Data_df["math_score"].mean()
Avg_math
```

```python
#Average Reading Score
Avg_read = Full_Data_df["reading_score"].mean()
Avg_read
```

```python
#% passing math (the percentage of students who passed math)
def math_result(row):
    if row["math_score"] < 70:
        return "Failed"
    else:
        return "Passed"
```

```python
Full_Data_df["math_result"] = Full_Data_df.apply(math_result, axis = 1)
Full_Data_df.head()
```

```python
math_p_count = Full_Data_df["math_result"].value_counts()["Passed"]
math_p_rate = math_p_count / Full_Data_df["math_score"].count()
math_p_rate
```

```python
#% passing reading (the percentage of students who passed reading)
def read_result(row):
    if row["reading_score"] < 70:
        return "Failed"
    else:
        return "Passed"
```

```python
Full_Data_df["reading_result"] = Full_Data_df.apply(read_result, axis = 1)
Full_Data_df.head()
```

```python
read_p_count = Full_Data_df["reading_result"].value_counts()["Passed"]
read_p_rate = read_p_count / Full_Data_df["reading_score"].count()
read_p_rate
```

```python
#% overall passing (the percentage of students who passed math AND reading)
def Passed_Both (row):
    if row["reading_result"] == "Passed" and row["math_result"] == "Passed":
        return "Yes"
    else:
        return "No"
```

```python
Full_Data_df["Passed_Both"] = Full_Data_df.apply(Passed_Both, axis = 1)
Full_Data_df.head()
```

```python
overall_p_count = Full_Data_df["Passed_Both"].value_counts()["Yes"]
overall_p_rate = read_p_count / Full_Data_df["Passed_Both"].count()
read_p_rate
```
